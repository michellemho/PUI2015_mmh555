```python
import plotly.plotly
from plotly.graph_objs import *
import pylab as pl
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
%pylab inline
```

```python
maleDF = pd.read_csv('M2015.csv')
femaleDF = pd.read_csv('F2015.csv')
```

```python
#femaleDF.head()
```

```python
#maleDF.head()
```

```python
maleDF['TimeParsed'] = pd.to_datetime(maleDF['Time'])
femaleDF['TimeParsed'] = pd.to_datetime(femaleDF['Time'])
#femaleDF['TimeParsed'].dt.hour[0].astype(int)
```

```python
mTotalSeconds = []
for i, n in enumerate(maleDF['TimeParsed']):    
    timeDelta = datetime.timedelta(hours = maleDF['TimeParsed'].dt.hour[i], minutes = maleDF['TimeParsed'].dt.minute[i], seconds = maleDF['TimeParsed'].dt.second[i])
    mTotalSeconds.append(timeDelta.total_seconds()/3600)
fTotalSeconds = []
for i, n in enumerate(femaleDF['TimeParsed']):    
    FtimeDelta = datetime.timedelta(hours = femaleDF['TimeParsed'].dt.hour[i].astype(int), minutes = femaleDF['TimeParsed'].dt.minute[i].astype(int), seconds = femaleDF['TimeParsed'].dt.second[i].astype(int))
    fTotalSeconds.append(FtimeDelta.total_seconds()/3600)
```

```python
FemaleTimes = pd.Series(fTotalSeconds, name='Female Time (hours)')
MaleTimes = pd.Series(mTotalSeconds, name='Male Time (hours)')

timeDF = pd.concat([FemaleTimes,MaleTimes],axis=1)
```

```python
marathonDF = timeDF.drop(100)
```

```python
pl.figure(figsize=(15,8))

sns.distplot(marathonDF['Male Time (hours)'], kde = False, bins =10, label = 'Male')
sns.distplot(marathonDF['Female Time (hours)'], kde = False, bins =10, label = 'Female')

pl.xlabel('Time (hours)', fontsize = 16)
pl.ylabel('Observations', fontsize = 16)
pl.legend(loc = 2)
pl.title('Comparing Distributions of the Top 100 Males and Females of the 2015 NYC Marathon')
```

```python
maleDF2014 = pd.read_csv('M2014.csv')
femaleDF2014 = pd.read_csv('F2014.csv')
```

```python
maleDF2014['TimeParsed'] = pd.to_datetime(maleDF2014['NetTime'])
femaleDF2014['TimeParsed'] = pd.to_datetime(femaleDF2014['NetTime'])

mTotalSeconds2014 = []
for i, n in enumerate(maleDF2014['TimeParsed']):    
    timeDelta = datetime.timedelta(hours = maleDF2014['TimeParsed'].dt.hour[i], minutes = maleDF2014['TimeParsed'].dt.minute[i], seconds = maleDF2014['TimeParsed'].dt.second[i])
    mTotalSeconds2014.append(timeDelta.total_seconds()/3600)
fTotalSeconds2014 = []
for i, n in enumerate(femaleDF2014['TimeParsed']):    
    FtimeDelta = datetime.timedelta(hours = femaleDF2014['TimeParsed'].dt.hour[i].astype(int), minutes = femaleDF2014['TimeParsed'].dt.minute[i].astype(int), seconds = femaleDF2014['TimeParsed'].dt.second[i].astype(int))
    fTotalSeconds2014.append(FtimeDelta.total_seconds()/3600)
    
FemaleTimes2014 = pd.Series(fTotalSeconds2014, name='Female Time (hours)')
MaleTimes2014 = pd.Series(mTotalSeconds2014, name='Male Time (hours)')

marathonDF2014 = pd.concat([FemaleTimes2014,MaleTimes2014],axis=1)
```

```python
pl.figure(figsize=(15,8))

sns.distplot(marathonDF2014['Male Time (hours)'], kde = False, bins =10, label = 'Male')
sns.distplot(marathonDF2014['Female Time (hours)'], kde = False, bins =10, label = 'Female')

pl.xlabel('Time (hours)', fontsize = 16)
pl.ylabel('Observations', fontsize = 16)
pl.legend(loc = 2)
pl.title('Comparing Distributions of the Top 100 Males and Females of the 2014 NYC Marathon')
```
